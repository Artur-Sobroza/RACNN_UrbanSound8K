```python
"""
Avalia√ß√£o da Generaliza√ß√£o do modelo RACNN pr√©-treinado no ESC-50 para o UrbanSound8K,
utilizando Transfer Learning (finetuning da cabe√ßa do modelo) e Valida√ß√£o Cruzada K-Fold.

Baseado no c√≥digo ESC-50_R11_GH.ipynb e adaptado para K-Fold e logging verboso.

"""

# BLOCO 1: Instala√ß√£o de bibliotecas e importa√ß√µes
# --------------------------------------------------------------------------------
# Estrat√©gia Detalhada - T√≥pico 1: Setup do Ambiente e Ferramentas
# --------------------------------------------------------------------------------
!pip install -q torchaudio librosa
!pip install -q einops # Necess√°rio para opera√ß√µes de reshape eficientes
!pip install -q scikit-learn # Necess√°rio para as novas m√©tricas de avalia√ß√£o
!pip install -q matplotlib # Garantir que matplotlib esteja instalado para plots
!pip install -q seaborn # Garantir que seaborn esteja instalado para plots

import os
import numpy as np
import pandas as pd
import torch
import torchaudio
import torch.nn as nn
import torch.nn.functional as F
import librosa # Mantido para compatibilidade
import librosa.display
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import StratifiedKFold # Usaremos StratifiedKFold para manter o balanceamento de classes
from tqdm import tqdm
from torchvision import transforms
from torchaudio.transforms import MelSpectrogram, AmplitudeToDB
from torch.optim.lr_scheduler import OneCycleLR
from google.colab import drive
import random
from einops import rearrange # Para opera√ß√µes de reshape complexas
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_recall_fscore_support
import seaborn as sns # Para visualiza√ß√£o da matriz de confus√£o
import time # Para medir o tempo de execu√ß√£o
import platform # Para informa√ß√µes do sistema
import sys # Para informa√ß√µes da vers√£o do Python
import pickle # <-- NOVO: Para salvar e carregar objetos Python

# BLOCO 1.1: Semente para reprodutibilidade
# --------------------------------------------------------------------------------
# Estrat√©gia Detalhada - T√≥pico 1: Setup do Ambiente e Ferramentas (Reprodutibilidade)
# --------------------------------------------------------------------------------
def seed_everything(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False # Define como False para reprodutibilidade
SEED = 42 # Semente global
seed_everything(SEED)

# Montar Google Drive
drive.mount('/content/drive')

# BLOCO 2: Defini√ß√£o de caminhos e hiperpar√¢metros
# --------------------------------------------------------------------------------
# Estrat√©gia Detalhada - T√≥pico 2: Prepara√ß√£o do Dataset UrbanSound8K
# Estrat√©gia Detalhada - T√≥pico 5: Arquitetura do Modelo RACNN (par√¢metros de entrada)
# --------------------------------------------------------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Caminhos para o Dataset UrbanSound8K
base_path_us8k = "/content/drive/MyDrive/Mestrado/Curso/1o Semestre 2025/DeepLearning/UrbanSound8K_dataset"
csv_path_us8k = os.path.join(base_path_us8k, "metadata", "UrbanSound8K.csv")
audio_base_path_us8k = os.path.join(base_path_us8k, "audio")
model_save_path_us8k = os.path.join(base_path_us8k, "modelos_finetune_US8K")
os.makedirs(model_save_path_us8k, exist_ok=True)

# Caminho para o modelo RACNN pr√©-treinado no ESC-50
PATH_TO_PRETRAINED_RACNN_ESC50 = "/content/drive/MyDrive/Mestrado/Curso/1o Semestre 2025/DeepLearning/ESC-50_dataset/modelos_ESC50/best_racnn_article_ESC50_fold2.pt" # SUBSTITUA PELO SEU MODELO ESC-50 REAL

# Hiperpar√¢metros para o finetuning no UrbanSound8K
SAMPLE_RATE = 44100 # Conforme artigo RACNN
N_FFT = 1024        # Conforme artigo RACNN
HOP_LENGTH = 512    # Conforme artigo RACNN
N_MELS = 128        # <---- Mantido em 128
DURATION_US8K = 4   # segundos (dura√ß√£o padronizada para UrbanSound8K)

# A Realidade do RACNN e Transfer Learning: Este √© o ponto crucial!
# O modelo RACNN espera um espectrograma de 216 frames, independentemente da dura√ß√£o do √°udio bruto
TARGET_MEL_FRAMES = 216

NUM_CLASSES_US8K = 10 # UrbanSound8K tem 10 classes

BATCH_SIZE_FINETUNE = 32 # Ajuste conforme a mem√≥ria da GPU
NUM_EPOCHS_FINETUNE = 50 # N√∫mero de √©pocas para o finetuning. EarlyStopping ajudar√° a parar antes
LEARNING_RATE_FINETUNE = 1e-4 # Taxa de aprendizado mais baixa para finetuning

# Estrat√©gia de divis√£o de dados para UrbanSound8K (valida√ß√£o cruzada)
N_SPLITS_KFOLD = 10 # 10 folds para valida√ß√£o cruzada

# Early Stopping
PATIENCE = 7 # Mantido em 7
MIN_DELTA = 0.001

# Valores MEAN/STD pr√©-calculados do ESC-50 para normaliza√ß√£o
# Manter estes valores para consist√™ncia com o pr√©-treinamento do RACNN
MEAN_ESC50 = -46.68
STD_ESC50 = 23.35


# BLOCO 3: Classe para Label Smoothing Cross Entropy (MANTIDO DO C√ìDIGO ESC-50_R11_GH.ipynb)
# --------------------------------------------------------------------------------
# Estrat√©gia Detalhada - T√≥pico 6: Crit√©rio de Perda (Loss Function)
# --------------------------------------------------------------------------------
class LabelSmoothingCrossEntropy(nn.Module):
    def __init__(self, epsilon: float = 0.1, reduction='mean'):
        super().__init__()
        self.epsilon = epsilon
        self.reduction = reduction

    def forward(self, preds, target):
        n = preds.size()[-1]
        log_preds = F.log_softmax(preds, dim=-1)
        loss = (-log_preds).sum(dim=-1)
        nll_loss = F.nll_loss(log_preds, target, reduction=self.reduction)
        return (1 - self.epsilon) * nll_loss + self.epsilon * loss.mean()

# BLOCO 4: Data Augmentation (SpecAugment) (MANTIDO DO C√ìDIGO ESC-50_R11_GH.ipynb)
# --------------------------------------------------------------------------------
# Estrat√©gia Detalhada - T√≥pico 4: Data Augmentation (SpecAugment)
# --------------------------------------------------------------------------------
class SpecAugment(nn.Module):
    def __init__(self, freq_mask_param=24, time_mask_param=48):
        super().__init__()
        self.spec_augment = nn.Sequential(
            torchaudio.transforms.FrequencyMasking(freq_mask_param),
            torchaudio.transforms.TimeMasking(time_mask_param),
        )

    def forward(self, x):
        return self.spec_augment(x)

# BLOCO 5: Defini√ß√£o do Dataset e Dataloader (MANTIDO DO C√ìDIGO ESC-50_R11_GH.ipynb)
# --------------------------------------------------------------------------------
# Estrat√©gia Detalhada - T√≥pico 3: Defini√ß√£o da Classe UrbanSound8KDataset (AudioDataset)
# --------------------------------------------------------------------------------
class AudioDataset(Dataset):
    def __init__(self, df, audio_base_dir, sample_rate, n_fft, hop_length, n_mels, duration,
                 target_mel_frames, transform=None, train=True, mean=None, std=None):
        self.annotations = df # DataFrame j√° filtrado para treino/valida√ß√£o
        self.audio_base_dir = audio_base_dir # Agora √© o diret√≥rio 'audio'
        self.sample_rate = sample_rate
        self.n_fft = n_fft
        self.hop_length = hop_length
        self.n_mels = n_mels
        self.duration = duration
        self.transform = transform
        self.train = train
        self.target_mel_frames = target_mel_frames # Definido para 216 frames

        # Inicializa as transforma√ß√µes de Mel Spectrogram apenas uma vez
        self.mel_spectrogram_transform = MelSpectrogram(
            sample_rate=self.sample_rate,
            n_fft=self.n_fft,
            hop_length=self.hop_length,
            n_mels=self.n_mels,
            normalized=False
        )
        self.amplitude_to_db_transform = AmplitudeToDB()

        # Valores MEAN/STD pr√©-calculados para ESC-50, usados para normalizar US8K para consist√™ncia
        self.mean = mean if mean is not None else MEAN_ESC50
        self.std = std if std is not None else STD_ESC50

    def __len__(self):
        return len(self.annotations)

    def __getitem__(self, index):
        # Para UrbanSound8K, o caminho completo inclui o 'fold'
        row = self.annotations.iloc[index]
        file_name = row['slice_file_name']
        fold_num = row['fold']
        label = row['classID'] # UrbanSound8K usa 'classID' como r√≥tulo

        audio_path = os.path.join(self.audio_base_dir, f"fold{fold_num}", file_name)

        waveform, sr = torchaudio.load(audio_path)

        # Resample se necess√°rio
        if sr != self.sample_rate:
            waveform = torchaudio.transforms.Resample(orig_freq=sr, new_freq=self.sample_rate)(waveform)
            sr = self.sample_rate

        # Padroniza√ß√£o da Dura√ß√£o do √Åudio (para 4 segundos)
        target_waveform_length = self.sample_rate * self.duration
        if waveform.shape[1] < target_waveform_length:
            padding = target_waveform_length - waveform.shape[1]
            waveform = F.pad(waveform, (0, padding))
        elif waveform.shape[1] > target_waveform_length:
            waveform = waveform[:, :target_waveform_length]

        # Converte para mono se for est√©reo
        if waveform.shape[0] > 1:
            waveform = torch.mean(waveform, dim=0, keepdim=True)

        # Gera√ß√£o do Mel-espectrograma
        mel_spec = self.mel_spectrogram_transform(waveform)
        mel_spec = self.amplitude_to_db_transform(mel_spec)

        # A Realidade do RACNN e Transfer Learning: Padroniza√ß√£o dos Frames do Espectrograma
        # Garante que o espectrograma tenha exatamente `self.target_mel_frames` (216) colunas
        if mel_spec.shape[2] < self.target_mel_frames:
            padding_frames = self.target_mel_frames - mel_spec.shape[2]
            mel_spec = F.pad(mel_spec, (0, padding_frames))
        elif mel_spec.shape[2] > self.target_mel_frames:
            mel_spec = mel_spec[:, :, :self.target_mel_frames]

        # Normaliza√ß√£o usando MEAN/STD do ESC-50
        mel_spec = (mel_spec - self.mean) / (self.std + 1e-6)

        # Data Augmentation (SpecAugment) - aplicado apenas se em modo de treino
        if self.transform and self.train:
            mel_spec = self.transform(mel_spec)

        # Garante que o label seja um inteiro e do tipo long
        return mel_spec, torch.tensor(int(label), dtype=torch.long)

# BLOCO 6: Implementa√ß√£o da Arquitetura RACNN (Conforme Artigo) (MANTIDO DO C√ìDIGO ESC-50_R11_GH.ipynb)
# --------------------------------------------------------------------------------
# Estrat√©gia Detalhada - T√≥pico 5: Arquitetura do Modelo RACNN
# --------------------------------------------------------------------------------
class RACModule(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(RACModule, self).__init__()
        self.dwconv = nn.Conv2d(in_channels, in_channels, kernel_size=3, padding=1, groups=in_channels, bias=False)
        self.bn_dw = nn.BatchNorm2d(in_channels)
        self.pwconv = nn.Conv2d(in_channels, out_channels, kernel_size=1, bias=False)
        self.bn_pw = nn.BatchNorm2d(out_channels)
        self.global_avg_pool = nn.AdaptiveAvgPool2d(1)
        self.fc1 = nn.Conv2d(out_channels, out_channels // 16, kernel_size=1, bias=False)
        self.relu = nn.ReLU(inplace=True)
        self.fc2 = nn.Conv2d(out_channels // 16, out_channels, kernel_size=1, bias=False)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.dwconv(x)
        x = self.bn_dw(x)
        x = F.relu(x, inplace=True)

        x = self.pwconv(x)
        x = self.bn_pw(x)
        out = F.relu(x, inplace=True)

        scale = self.global_avg_pool(out)
        scale = self.fc1(scale)
        scale = self.relu(scale)
        scale = self.fc2(scale)
        scale = self.sigmoid(scale)

        out = out * scale.expand_as(out)
        return out

class RACBlock(nn.Module):
    def __init__(self, in_channels, out_channels, stride=1, num_rac_modules=1):
        super(RACBlock, self).__init__()
        layers = []
        layers.append(RACModule(in_channels, out_channels))
        for _ in range(1, num_rac_modules):
            layers.append(RACModule(out_channels, out_channels))
        self.rac_modules = nn.Sequential(*layers)

        if stride == 2:
            self.pool = nn.AvgPool2d(kernel_size=2, stride=2)
        else:
            self.pool = nn.Identity()

    def forward(self, x):
        x = self.rac_modules(x)
        x = self.pool(x)
        return x

class RACNN(nn.Module):
    def __init__(self, num_classes=50): # Inicializado com 50 classes para carregar pesos do ESC-50
        super(RACNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(32)
        self.relu1 = nn.ReLU(inplace=True)
        self.pool1 = nn.AvgPool2d(kernel_size=2, stride=2)

        self.rac_block2 = nn.Sequential(
            RACModule(32, 64),
            RACModule(64, 64)
        )
        self.pool2 = nn.AvgPool2d(kernel_size=2, stride=2)

        self.rac_block3 = nn.Sequential(
            RACModule(64, 128),
            RACModule(128, 128)
        )
        self.pool3 = nn.AvgPool2d(kernel_size=2, stride=2)

        self.rac_block4 = nn.Sequential(
            RACModule(128, 256),
            RACModule(256, 256)
        )
        self.pool4 = nn.AvgPool2d(kernel_size=2, stride=2)

        self.global_avg_pool = nn.AdaptiveAvgPool2d(1)

        self.fc1 = nn.Linear(256, 512)
        self.bn_fc1 = nn.BatchNorm1d(512)
        self.dropout1 = nn.Dropout(0.5)

        self.fc2 = nn.Linear(512, num_classes) # Esta ser√° a camada de sa√≠da a ser substitu√≠da

    def forward(self, x):
        x = self.conv1(x)
        x = self.bn1(x)
        x = self.relu1(x)
        x = self.pool1(x)

        x = self.rac_block2(x)
        x = self.pool2(x)

        x = self.rac_block3(x)
        x = self.pool3(x)

        x = self.rac_block4(x)
        x = self.pool4(x)

        x = self.global_avg_pool(x)
        x = torch.flatten(x, 1)

        x = self.fc1(x)
        x = self.bn_fc1(x)
        x = F.relu(x, inplace=True)
        x = self.dropout1(x)

        x = self.fc2(x)

        return x

    # M√©todos de Transfer Learning
    # --------------------------------------------------------------------------------
    # Estrat√©gia Detalhada - T√≥pico 5: Arquitetura do Modelo RACNN (M√©todos de Transfer Learning)
    # --------------------------------------------------------------------------------
    def freeze_except_head(self):
        """
        Congela todas as camadas do modelo, exceto as camadas de classifica√ß√£o (a "cabe√ßa").
        """
        for param in self.parameters():
            param.requires_grad = False

        # Descongelar as camadas da cabe√ßa para finetuning
        for param in self.fc1.parameters():
            param.requires_grad = True
        for param in self.bn_fc1.parameters():
            param.requires_grad = True
        for param in self.fc2.parameters():
            param.requires_grad = True
        print("‚úÖ Camadas do modelo congeladas, exceto a cabe√ßa (fc1, bn_fc1, fc2).")


    def replace_head(self, new_num_classes):
        """
        Substitui a camada de sa√≠da (fc2) para o n√∫mero de classes do novo dataset.
        """
        old_fc2_in_features = self.fc2.in_features
        self.fc2 = nn.Linear(old_fc2_in_features, new_num_classes)
        # Os par√¢metros da nova camada s√£o automaticamente trein√°veis por padr√£o
        print(f"‚úÖ Camada de sa√≠da (fc2) substitu√≠da para {new_num_classes} classes.")

# BLOCO 7: Fun√ß√µes Auxiliares de Treinamento e Valida√ß√£o (MANTIDO DO C√ìDIGO ESC-50_R11_GH.ipynb)
# --------------------------------------------------------------------------------
# Estrat√©gia Detalhada - T√≥pico 7: Fun√ß√µes de Treinamento e Avalia√ß√£o
# --------------------------------------------------------------------------------
class EarlyStopper:
    def __init__(self, patience=5, min_delta=0):
        super().__init__()
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.min_validation_acc = 0.0 # Monitoramos a acur√°cia de valida√ß√£o
        self.early_stop = False

    def step(self, validation_acc):
        if validation_acc > self.min_validation_acc + self.min_delta:
            self.min_validation_acc = validation_acc
            self.counter = 0 # Reseta o contador se houver melhoria
        else:
            self.counter += 1 # Incrementa se n√£o houver melhoria significativa
            if self.counter >= self.patience:
                self.early_stop = True
                print(f"Paci√™ncia esgotada ({self.patience} √©pocas sem melhoria).")

def train_epoch(model, dataloader, criterion, optimizer, scheduler, device, epoch_num, total_epochs):
    model.train()
    total_loss = 0
    correct = 0
    total = 0
    progress_bar = tqdm(dataloader, desc=f"üìò Treino - √âpoca {epoch_num}/{total_epochs}")

    for batch_idx, (X, y) in enumerate(progress_bar):
        X, y = X.to(device), y.to(device).long()

        optimizer.zero_grad()
        outputs = model(X)
        loss = criterion(outputs, y)
        loss.backward()
        optimizer.step()
        scheduler.step() # Atualiza o scheduler a cada batch

        total_loss += loss.item() * X.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total += y.size(0)
        correct += (predicted == y).sum().item()

        current_lr = optimizer.param_groups[0]['lr']
        progress_bar.set_postfix(loss=loss.item(), acc=correct/total, lr=f"{current_lr:.6f}")

    train_epoch_loss = total_loss / total
    train_epoch_acc = correct / total
    return train_epoch_loss, train_epoch_acc

def validate_epoch(model, dataloader, criterion, device):
    model.eval()
    val_loss = 0
    val_correct = 0
    val_total = 0
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for X, y in dataloader:
            X, y = X.to(device), y.to(device).long()
            outputs = model(X)
            loss = criterion(outputs, y)

            val_loss += loss.item() * X.size(0)
            _, preds = torch.max(outputs, 1)
            val_correct += (preds == y).sum().item()
            val_total += y.size(0)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(y.cpu().numpy())

    val_epoch_loss = val_loss / val_total
    val_epoch_acc = val_correct / val_total
    return val_epoch_loss, val_epoch_acc, all_preds, all_labels

# BLOCO 8: Processo de Finetuning Principal (ADAPTADO PARA K-FOLD E LOGGING)
# --------------------------------------------------------------------------------
# Estrat√©gia Detalhada - T√≥pico 8: Estrat√©gia de Divis√£o de Dados (K-Fold para UrbanSound8K)
# Estrat√©gia Detalhada - T√≥pico 9: Processo de Finetuning
# Estrat√©gia Detalhada - T√≥pico 10: Sa√≠da Esperada
# --------------------------------------------------------------------------------

print("\n====================================================================================")
print("                           INICIANDO EXPERIMENTO K-FOLD FINETUNING RACNN NO URBANOUND8K")
print("====================================================================================")

# --------------------------------------------------------------------------------
# LOGGING: Configura√ß√£o do Ambiente
# --------------------------------------------------------------------------------
print("\nüìù Configura√ß√µes do Ambiente:")
print(f"  - Sistema Operacional: {platform.system()} {platform.release()}")
print(f"  - Python Vers√£o: {sys.version}")
print(f"  - PyTorch Vers√£o: {torch.__version__}")
print(f"  - Torchaudio Vers√£o: {torchaudio.__version__}")
print(f"  - Librosa Vers√£o: {librosa.__version__}")
print(f"  - Pandas Vers√£o: {pd.__version__}")
print(f"  - Scikit-learn Vers√£o: {os.environ.get('SCIKIT_LEARN_VERSION', 'N√£o especificado')}") # scikit-learn n√£o tem __version__ direto
print(f"  - Dispositivo Selecionado: {device}")
if torch.cuda.is_available():
    print(f"  - GPU Nome: {torch.cuda.get_device_name(0)}")
    print(f"  - Mem√≥ria Total da GPU: {torch.cuda.get_device_properties(0).total_memory / (1024**3):.2f} GB")
print(f"  - Semente de Reprodutibilidade: {SEED}")
print(f"  - Caminho Base UrbanSound8K: {base_path_us8k}")
print(f"  - Caminho CSV UrbanSound8K: {csv_path_us8k}")
print(f"  - Caminho √Åudio UrbanSound8K: {audio_base_path_us8k}")
print(f"  - Caminho Salvar Modelos: {model_save_path_us8k}")
print(f"  - Caminho Modelo RACNN ESC-50 Pr√©-treinado: {PATH_TO_PRETRAINED_RACNN_ESC50}")

print("\n‚öôÔ∏è Hiperpar√¢metros do Finetuning:")
print(f"  - Sample Rate: {SAMPLE_RATE}")
print(f"  - N_FFT: {N_FFT}")
print(f"  - Hop Length: {HOP_LENGTH}")
print(f"  - N_MELS: {N_MELS}")
print(f"  - Target Mel Frames: {TARGET_MEL_FRAMES}")
print(f"  - Dura√ß√£o do √Åudio: {DURATION_US8K} segundos")
print(f"  - N√∫mero de Classes (UrbanSound8K): {NUM_CLASSES_US8K}")
print(f"  - Batch Size Finetune: {BATCH_SIZE_FINETUNE}")
print(f"  - Num Workers (DataLoader): 2")
print(f"  - N√∫mero M√°ximo de √âpocas: {NUM_EPOCHS_FINETUNE}")
print(f"  - Learning Rate Finetune: {LEARNING_RATE_FINETUNE}")
print(f"  - Paci√™ncia Early Stopping: {PATIENCE}")
print(f"  - Delta M√≠nimo Early Stopping: 0.001")
print(f"  - Normaliza√ß√£o ESC-50 (Mean/Std): {MEAN_ESC50:.2f} / {STD_ESC50:.2f}")
print(f"  - N√∫mero de Folds K-Fold: {N_SPLITS_KFOLD}")
print("------------------------------------------------------------------------------------")


# Carregar o dataset completo do UrbanSound8K
print(f"\n‚è≥ Carregando metadados do UrbanSound8K de: {csv_path_us8k}")
try:
    us8k_df = pd.read_csv(csv_path_us8k)
    print(f"‚úÖ Metadados carregados. Total de amostras: {len(us8k_df)}")
    print(f"N√∫mero de classes √∫nicas no UrbanSound8K: {us8k_df['classID'].nunique()}")
except FileNotFoundError:
    print(f"‚ùå ERRO: Arquivo CSV do UrbanSound8K n√£o encontrado em {csv_path_us8k}")
    print("Por favor, verifique o caminho e a montagem do Google Drive.")
    exit() # Aborta a execu√ß√£o se o CSV n√£o for encontrado

# Mapeamento de classID para nome da classe para relat√≥rios
class_names_us8k = sorted(us8k_df['class'].unique())
class_id_to_name = {id: name for id, name in enumerate(class_names_us8k)}
print(f"Nomes das classes UrbanSound8K: {class_names_us8k}")

# -------- VISUALIZA√á√ÉO DO BALANCEAMENTO DE CLASSES (MANTIDO DO C√ìDIGO ESC-50_R11_GH.ipynb) --------
print("\nüìä An√°lise do Balanceamento de Classes no UrbanSound8K (Total):")
class_counts_total = us8k_df['class'].value_counts().sort_index()
print(class_counts_total)

plt.figure(figsize=(12, 6))
# A√á√ÉO: Ajuste para o FutureWarning do seaborn
sns.barplot(x=class_counts_total.index, y=class_counts_total.values, palette='viridis', hue=class_counts_total.index, legend=False)
plt.title('Distribui√ß√£o de Amostras por Classe (UrbanSound8K - Total)')
plt.xlabel('Classe')
plt.ylabel('N√∫mero de Amostras')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
# -------- FIM DA VISUALIZA√á√ÉO --------

# --- NOVA L√ìGICA: SALVAR E CARREGAR M√âTRICAS DOS FOLDS ANTERIORES ---
# <----- IN√çCIO DA NOVA L√ìGICA ----->
metrics_history_file = os.path.join(model_save_path_us8k, "kfold_metrics_history.pkl")

# Inicializar listas para armazenar resultados de todos os folds
all_fold_metrics = []
all_fold_classification_reports = []
all_fold_confusion_matrices = []
all_fold_times = [] # all_fold_best_epochs ser√° extra√≠do de all_fold_metrics

# Tentar carregar hist√≥rico de execu√ß√µes anteriores
if os.path.exists(metrics_history_file):
    print(f"\nüîÑ Carregando hist√≥rico de m√©tricas de folds anteriores de: {metrics_history_file}")
    try:
        with open(metrics_history_file, 'rb') as f:
            loaded_data = pickle.load(f)
            all_fold_metrics = loaded_data.get('all_fold_metrics', [])
            all_fold_classification_reports = loaded_data.get('all_fold_classification_reports', [])
            all_fold_confusion_matrices = loaded_data.get('all_fold_confusion_matrices', [])
            all_fold_times = loaded_data.get('all_fold_times', [])
        print(f"‚úÖ Hist√≥rico carregado. {len(all_fold_metrics)} folds j√° foram processados.")
    except Exception as e:
        print(f"‚ùå ERRO ao carregar hist√≥rico de m√©tricas: {e}. Iniciando com listas vazias.")
        all_fold_metrics = []
        all_fold_classification_reports = []
        all_fold_confusion_matrices = []
        all_fold_times = []
else:
    print(f"\n‚ÑπÔ∏è Nenhum hist√≥rico de m√©tricas encontrado em {metrics_history_file}. Iniciando experimento do zero.")

# Determinar o fold inicial com base no hist√≥rico carregado
# Se j√° processamos N folds, o pr√≥ximo fold a ser executado √© N+1.
start_fold_idx = len(all_fold_metrics) + 1
# Certifica-se de que n√£o estamos tentando iniciar um fold que j√° est√° al√©m do total
if start_fold_idx > N_SPLITS_KFOLD:
    print("‚úÖ Todos os folds j√° foram processados na execu√ß√£o anterior. Resultados finais podem ser gerados.")
    # Se todos os folds j√° foram processados, podemos pular o loop de treino e ir direto para a an√°lise final.
    # Definindo start_fold_idx para um valor que far√° o loop principal n√£o executar.
    start_fold_idx = N_SPLITS_KFOLD + 1
# <----- FIM DA NOVA L√ìGICA ----->

# K-Fold Cross-Validation Setup
# Usamos StratifiedKFold para garantir que cada fold tenha uma distribui√ß√£o de classes similar
skf = StratifiedKFold(n_splits=N_SPLITS_KFOLD, shuffle=True, random_state=SEED)

print(f"\nüöÄ Iniciando K-Fold Cross-Validation com {N_SPLITS_KFOLD} Folds...")
total_experiment_start_time = time.time()

# Iterar sobre os folds
# us8k_df['fold'] j√° cont√©m os n√∫meros dos folds originais, podemos us√°-los diretamente
# para criar os splits com StratifiedKFold, garantindo a estratifica√ß√£o baseada nas classes
# para cada itera√ß√£o, onde um 'fold' √© o conjunto de valida√ß√£o e os demais s√£o o treino.

# A estrat√©gia do UrbanSound8K j√° define os folds de 1 a 10. Podemos iterar por eles.
# Loop original: for fold_idx in range(1, N_SPLITS_KFOLD + 1):
# <----- LINHA ALTERADA PARA RECUPERA√á√ÉO DE EXECU√á√ÉO ----->
for fold_idx in range(start_fold_idx, N_SPLITS_KFOLD + 1):
# <----- FIM DA LINHA ALTERADA ----->
    print(f"\n{'='*20} INICIANDO FOLD {fold_idx}/{N_SPLITS_KFOLD} {'='*20}")
    fold_start_time = time.time()

    # Dividir o dataset com base no fold atual para valida√ß√£o
    train_df = us8k_df[us8k_df['fold'] != fold_idx]
    val_df = us8k_df[us8k_df['fold'] == fold_idx]

    print(f"üìä Divis√£o de Dados para Fold {fold_idx}:")
    print(f"  - Amostras de Treino: {len(train_df)}")
    print(f"  - Amostras de Valida√ß√£o: {len(val_df)}")

    # Contagem de Amostras por Classe (Treino e Valida√ß√£o do Fold Atual)
    print("\n  - Contagem de Amostras por Classe (Conjunto de TREINO do Fold Atual):")
    class_counts_train_fold = train_df['class'].value_counts().sort_index()
    print(class_counts_train_fold)

    print("\n  - Contagem de Amostras por Classe (Conjunto de VALIDA√á√ÉO do Fold Atual):")
    class_counts_val_fold = val_df['class'].value_counts().sort_index()
    print(class_counts_val_fold)


    # Instanciar o SpecAugment para o treino (com par√¢metros do artigo)
    train_transform = SpecAugment(freq_mask_param=24, time_mask_param=48)

    # Criar inst√¢ncias do AudioDataset
    train_dataset = AudioDataset(
        df=train_df,
        audio_base_dir=audio_base_path_us8k,
        sample_rate=SAMPLE_RATE,
        n_fft=N_FFT,
        hop_length=HOP_LENGTH,
        n_mels=N_MELS,
        duration=DURATION_US8K,
        target_mel_frames=TARGET_MEL_FRAMES,
        transform=train_transform,
        train=True,
        mean=MEAN_ESC50,
        std=STD_ESC50
    )

    val_dataset = AudioDataset(
        df=val_df,
        audio_base_dir=audio_base_path_us8k,
        sample_rate=SAMPLE_RATE,
        n_fft=N_FFT,
        hop_length=HOP_LENGTH,
        n_mels=N_MELS,
        duration=DURATION_US8K,
        target_mel_frames=TARGET_MEL_FRAMES,
        transform=None, # Sem SpecAugment para valida√ß√£o
        train=False,
        mean=MEAN_ESC50,
        std=STD_ESC50
    )

    # Criar DataLoaders
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE_FINETUNE, shuffle=True, num_workers=2, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE_FINETUNE, shuffle=False, num_workers=2, pin_memory=True)

    # --------------------------------------------------------------------------------
    # Estrat√©gia Detalhada - T√≥pico 9: Processo de Finetuning (Carregar Modelo Pr√©-treinado)
    # --------------------------------------------------------------------------------
    # Carregar o modelo RACNN pr√©-treinado no ESC-50
    print(f"\nüß† Carregando modelo RACNN pr√©-treinado do ESC-50 para Fold {fold_idx}: {PATH_TO_PRETRAINED_RACNN_ESC50}")
    model = RACNN(num_classes=50).to(device) # Inicializa com 50 classes para carregar os pesos do ESC-50

    try:
        model.load_state_dict(torch.load(PATH_TO_PRETRAINED_RACNN_ESC50, map_location=device))
        print("‚úÖ Modelo pr√©-treinado carregado com sucesso!")
    except FileNotFoundError:
        print(f"‚ùå ERRO: Modelo pr√©-treinado n√£o encontrado em {PATH_TO_PRETRAINED_RACNN_ESC50}")
        print("Por favor, verifique o caminho do modelo ESC-50 salvo e tente novamente.")
        exit() # Aborta a execu√ß√£o se o modelo pr√©-treinado n√£o for encontrado
    except Exception as e:
        print(f"‚ùå ERRO ao carregar o modelo pr√©-treinado: {e}")
        print("Verifique se o modelo est√° salvo corretamente e √© compat√≠vel com a arquitetura RACNN.")
        exit()

    # --------------------------------------------------------------------------------
    # Estrat√©gia Detalhada - T√≥pico 9: Processo de Finetuning (Congelamento e Substitui√ß√£o da Cabe√ßa)
    # --------------------------------------------------------------------------------
    # Congelar camadas e substituir a cabe√ßa do modelo
    model.freeze_except_head()
    model.replace_head(NUM_CLASSES_US8K) # Substitui a camada de sa√≠da para 10 classes do UrbanSound8K
    model.to(device) # Garante que o modelo esteja no dispositivo correto ap√≥s modifica√ß√µes

    # --------------------------------------------------------------------------------
    # Estrat√©gia Detalhada - T√≥pico 9: Processo de Finetuning (Otimizador e Scheduler)
    # --------------------------------------------------------------------------------
    # Otimizador: Apenas para par√¢metros que 'requires_grad=True' (cabe√ßa do modelo)
    optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=LEARNING_RATE_FINETUNE)
    criterion = LabelSmoothingCrossEntropy(epsilon=0.1)

    # Scheduler OneCycleLR
    scheduler = OneCycleLR(optimizer,
                           max_lr=LEARNING_RATE_FINETUNE * 10, # Max LR pode ser maior para OneCycleLR
                           steps_per_epoch=len(train_loader),
                           epochs=NUM_EPOCHS_FINETUNE,
                           anneal_strategy='cos',
                           pct_start=0.3,
                           div_factor=25.0,
                           final_div_factor=10000.0)

    early_stopper = EarlyStopper(patience=PATIENCE, min_delta=MIN_DELTA)

    # Listas para armazenar o hist√≥rico de treinamento do fold atual (n√£o s√£o as globais)
    train_loss_history = []
    train_acc_history = []
    val_loss_history = []
    val_acc_history = []

    # Vari√°veis para armazenar o melhor desempenho do fold atual
    best_val_acc_fold = 0.0
    best_val_loss_fold = float('inf')
    best_epoch_fold = 0

    best_model_path_fold = os.path.join(model_save_path_us8k, f"best_racnn_finetune_US8K_fold{fold_idx}.pt")

    # Vari√°veis para armazenar as previs√µes e r√≥tulos do melhor modelo do fold atual
    best_val_preds_fold = []
    best_val_labels_fold = []

    print(f"\nüöÄ Iniciando o Treinamento/Finetuning para o Fold {fold_idx}...")

    for epoch in range(1, NUM_EPOCHS_FINETUNE + 1):
        # Treino
        train_epoch_loss, train_epoch_acc = train_epoch(model, train_loader, criterion, optimizer, scheduler, device, epoch, NUM_EPOCHS_FINETUNE)
        train_loss_history.append(train_epoch_loss)
        train_acc_history.append(train_epoch_acc)

        # Valida√ß√£o
        val_epoch_loss, val_epoch_acc, current_val_preds, current_val_labels = validate_epoch(model, val_loader, criterion, device)
        val_loss_history.append(val_epoch_loss)
        val_acc_history.append(val_epoch_acc)

        current_lr = optimizer.param_groups[0]['lr']
        log_message = (
            f"‚úÖ √âpoca {epoch}/{NUM_EPOCHS_FINETUNE} | Treino Loss: {train_epoch_loss:.4f} | Treino Acur√°cia: {train_epoch_acc:.4f} "
            f"| Valida√ß√£o Loss: {val_epoch_loss:.4f} | Valida√ß√£o Acur√°cia: {val_epoch_acc:.4f} "
            f"| LR: {current_lr:.6f}"
        )

        # Condi√ß√£o para salvar o melhor modelo: baseada na acur√°cia de valida√ß√£o
        if val_epoch_acc > best_val_acc_fold:
            best_val_acc_fold = val_epoch_acc
            best_val_loss_fold = val_epoch_loss
            best_epoch_fold = epoch
            torch.save(model.state_dict(), best_model_path_fold)
            best_val_preds_fold = current_val_preds
            best_val_labels_fold = current_val_labels
            log_message += f" | üèÜ Novo Melhor Modelo Salvo (Acur√°cia: {best_val_acc_fold:.4f}, Loss: {best_val_loss_fold:.4f})!"
        print(log_message)

        # Early Stopping
        early_stopper.step(val_epoch_acc)
        if early_stopper.early_stop:
            print(f"\n‚èπÔ∏è Early stopping ativado para o Fold {fold_idx}. Melhor Acur√°cia de Valida√ß√£o: {best_val_acc_fold:.4f} (√âpoca: {best_epoch_fold})")
            break

    fold_end_time = time.time()
    fold_total_time = fold_end_time - fold_start_time

    print(f"\n{'='*15} RESULTADOS DO FOLD {fold_idx} {'='*15}")
    print(f"  - Tempo de execu√ß√£o do Fold {fold_idx}: {fold_total_time/60:.2f} minutos ({fold_total_time:.2f} segundos)")
    print(f"  - Melhor Acur√°cia de Valida√ß√£o (Fold {fold_idx}): {best_val_acc_fold:.4f}")
    print(f"  - Melhor Loss de Valida√ß√£o (Fold {fold_idx}): {best_val_loss_fold:.4f}")
    print(f"  - Melhor Modelo Salvo na √âpoca: {best_epoch_fold}")
    print(f"  - Caminho do Melhor Modelo Salvo: {best_model_path_fold}")

    # <---- AN√ÅLISE DE M√âTRICAS DETALHADAS PARA O FOLD ATUAL
    print(f"\nüìä An√°lise Detalhada do Melhor Modelo para o Fold {fold_idx}:")

    # Relat√≥rio de Classifica√ß√£o
    target_names_us8k = [class_id_to_name[i] for i in sorted(class_id_to_name.keys())]
    report = classification_report(best_val_labels_fold, best_val_preds_fold, target_names=target_names_us8k, output_dict=True)

    print("\n--- Relat√≥rio de Classifica√ß√£o ---")
    print(classification_report(best_val_labels_fold, best_val_preds_fold, target_names=target_names_us8k))

    # Matriz de Confus√£o
    cm_fold = confusion_matrix(best_val_labels_fold, best_val_preds_fold)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm_fold, annot=True, fmt='d', cmap='Blues', xticklabels=target_names_us8k, yticklabels=target_names_us8k)
    plt.xlabel('Previs√£o')
    plt.ylabel('Verdadeiro')
    plt.title(f'Matriz de Confus√£o do Melhor Modelo (Fold {fold_idx} Valida√ß√£o)')
    plt.show()

    # <----- IN√çCIO DA NOVA L√ìGICA: ARMAZENAR E SALVAR M√âTRICAS DO FOLD ATUAL ----->
    # Armazenar m√©tricas do fold atual para agrega√ß√£o futura
    all_fold_metrics.append({
        'fold': fold_idx,
        'best_accuracy': best_val_acc_fold,
        'best_loss': best_val_loss_fold,
        'best_epoch': best_epoch_fold,
        'time_seconds': fold_total_time,
    })
    all_fold_classification_reports.append(report) # Armazena o relat√≥rio completo
    all_fold_confusion_matrices.append(cm_fold) # Armazena a matriz de confus√£o
    all_fold_times.append(fold_total_time) # Armazena o tempo do fold

    # Salvar o estado atual das m√©tricas ap√≥s cada fold
    try:
        with open(metrics_history_file, 'wb') as f:
            pickle.dump({
                'all_fold_metrics': all_fold_metrics,
                'all_fold_classification_reports': all_fold_classification_reports,
                'all_fold_confusion_matrices': all_fold_confusion_matrices,
                'all_fold_times': all_fold_times
            }, f)
        print(f"üíæ Hist√≥rico de m√©tricas salvo para {len(all_fold_metrics)} folds em: {metrics_history_file}")
    except Exception as e:
        print(f"‚ùå ERRO ao salvar hist√≥rico de m√©tricas: {e}")
    # <----- FIM DA NOVA L√ìGICA ----->

    early_stopper.early_stop = False # Resetar early stopper para o pr√≥ximo fold

total_experiment_end_time = time.time()
# Calcular o tempo total do experimento apenas se ele foi iniciado nesta execu√ß√£o
if start_fold_idx == 1:
    total_experiment_time = total_experiment_end_time - total_experiment_start_time
else:
    # Se carregou hist√≥rico, somamos os tempos antigos com os novos
    total_experiment_time = sum(all_fold_times) # all_fold_times j√° tem todos os tempos se carregado/adicionado

total_experiment_time_minutes = total_experiment_time / 60


print("\n====================================================================================")
print("                           PROCESSO K-FOLD CONCLU√çDO")
print("====================================================================================")
print(f"‚ú® Tempo Total do Experimento K-Fold: {total_experiment_time_minutes:.2f} minutos ({total_experiment_time:.2f} segundos)")

# BLOCO 9: Resultados Finais Agregados (NOVO)
# --------------------------------------------------------------------------------
# Estrat√©gia Detalhada - T√≥pico 10: Sa√≠da Esperada (M√©dias e Desvios Padr√£o)
# --------------------------------------------------------------------------------
print("\nüìä RESULTADOS FINAIS AGREGADOS (M√©dia e Desvio Padr√£o sobre 10 Folds):")

# Coletar m√©tricas de todos os folds
# all_fold_metrics agora j√° deve conter todos os folds (se carregados e/ou processados)
accuracies = [m['best_accuracy'] for m in all_fold_metrics]
losses = [m['best_loss'] for m in all_fold_metrics]
epochs_reached = [m['best_epoch'] for m in all_fold_metrics]

# Calcular m√©dias e desvios padr√£o para m√©tricas gerais
mean_accuracy = np.mean(accuracies)
std_accuracy = np.std(accuracies)
mean_loss = np.mean(losses)
std_loss = np.std(losses)
mean_epochs = np.mean(epochs_reached)
std_epochs = np.std(epochs_reached)

print(f"\n  - Acur√°cia M√©dia: {mean_accuracy:.4f} ¬± {std_accuracy:.4f}")
print(f"  - Loss M√©dia: {mean_loss:.4f} ¬± {std_loss:.4f}")
print(f"  - √âpocas M√©dia Alcan√ßada (Melhor Modelo): {mean_epochs:.2f} ¬± {std_epochs:.2f}")
# all_fold_times j√° cont√©m os tempos de todos os folds, somados ao carregar.
print(f"  - Tempo M√©dio por Fold: {np.mean(all_fold_times)/60:.2f} minutos ({np.mean(all_fold_times):.2f} segundos)")


# Agrega√ß√£o das m√©tricas do Classification Report (Precision, Recall, F1-score)
# Para macro avg, weighted avg e por classe
metrics_to_aggregate = ['accuracy', 'macro avg', 'weighted avg'] + target_names_us8k

aggregated_report_data = {}
for metric_type in metrics_to_aggregate:
    aggregated_report_data[metric_type] = {'precision': [], 'recall': [], 'f1-score': []}

for report in all_fold_classification_reports:
    # Acur√°cia √© um campo direto no dicion√°rio do report
    if 'accuracy' in report:
        aggregated_report_data['accuracy']['f1-score'].append(report['accuracy']) # Usamos f1-score para armazenar a acur√°cia para consist√™ncia no loop

    for class_name in target_names_us8k:
        if class_name in report:
            aggregated_report_data[class_name]['precision'].append(report[class_name]['precision'])
            aggregated_report_data[class_name]['recall'].append(report[class_name]['recall'])
            aggregated_report_data[class_name]['f1-score'].append(report[class_name]['f1-score'])

    # Adicionar macro avg e weighted avg
    if 'macro avg' in report:
        aggregated_report_data['macro avg']['precision'].append(report['macro avg']['precision'])
        aggregated_report_data['macro avg']['recall'].append(report['macro avg']['recall'])
        aggregated_report_data['macro avg']['f1-score'].append(report['macro avg']['f1-score'])
    if 'weighted avg' in report:
        aggregated_report_data['weighted avg']['precision'].append(report['weighted avg']['precision'])
        aggregated_report_data['weighted avg']['recall'].append(report['weighted avg']['recall'])
        aggregated_report_data['weighted avg']['f1-score'].append(report['weighted avg']['f1-score'])


print("\n--- Relat√≥rio de Classifica√ß√£o M√©dio (10 Folds) ---")
# Cabe√ßalho da tabela
print(f"{'Classe':<20} | {'Precision':<15} | {'Recall':<15} | {'F1-Score':<15}")
print(f"{'-'*20}-+-{'-'*15}-+-{'-'*15}-+-{'-'*15}")

# Imprimir m√©tricas por classe e m√©dias agregadas
for key in target_names_us8k + ['macro avg', 'weighted avg']:
    if key == 'accuracy': # Acur√°cia j√° foi tratada separadamente, mas para garantir
        continue
    if key in aggregated_report_data and aggregated_report_data[key]['precision']:
        mean_precision = np.mean(aggregated_report_data[key]['precision'])
        std_precision = np.std(aggregated_report_data[key]['precision'])
        mean_recall = np.mean(aggregated_report_data[key]['recall'])
        std_recall = np.std(aggregated_report_data[key]['recall'])
        mean_f1_score = np.mean(aggregated_report_data[key]['f1-score'])
        std_f1_score = np.std(aggregated_report_data[key]['f1-score'])

        print(f"{key:<20} | {mean_precision:.4f} ¬± {std_precision:.4f} | {mean_recall:.4f} ¬± {std_recall:.4f} | {mean_f1_score:.4f} ¬± {std_f1_score:.4f}")
    else:
        print(f"{key:<20} | {'N/A':<15} | {'N/A':<15} | {'N/A':<15}") # Caso a classe n√£o tenha sido encontrada em algum fold

# M√©dia da matriz de confus√£o
if all_fold_confusion_matrices:
    # Garante que todas as matrizes tenham o mesmo shape antes de calcular a m√©dia
    # (Pode ser um problema se houve algum erro inesperado em folds anteriores e n√£o foram coletadas)
    if all_fold_confusion_matrices: # Verifica se a lista n√£o est√° vazia
        # Pega o shape da primeira matriz e verifica se as demais s√£o compat√≠veis
        ref_shape = all_fold_confusion_matrices[0].shape
        valid_matrices = [cm for cm in all_fold_confusion_matrices if cm.shape == ref_shape]

        if valid_matrices:
            mean_cm = np.mean(valid_matrices, axis=0).astype(int)
            print("\n--- Matriz de Confus√£o M√©dia (Arredondada para Inteiro) ---")
            plt.figure(figsize=(10, 8))
            sns.heatmap(mean_cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_names_us8k, yticklabels=target_names_us8k)
            plt.xlabel('Previs√£o')
            plt.ylabel('Verdadeiro')
            plt.title('Matriz de Confus√£o M√©dia (10 Folds)')
            plt.show()
        else:
            print("\nN√£o foi poss√≠vel gerar a matriz de confus√£o m√©dia (matrizes com shapes inconsistentes ou vazias).")
    else:
        print("\nN√£o foi poss√≠vel gerar a matriz de confus√£o m√©dia (nenhuma matriz coletada).")

print(f"\n‚ú® Processo de Finetuning K-Fold para UrbanSound8K conclu√≠do!")
```
