```python
"""
Implementação da arquitetura RACNN para ESC-50, baseada no artigo:
"Fast environmental sound classification based on resource adaptive convolutional neural network"
(Zheng Fang, Bo Yin, Zehua Du, Xianqing Huang, Scientific Reports, 2022).

Incorpora as melhorias do ESC-50_R9 (OneCycleLR, LabelSmoothingCrossEntropy, normalização, seed),
implementação de K-Fold Cross-Validation e logs de treinamento aprimorados.
"""

# BLOCO 1: Instalação de bibliotecas e importações
!pip install -q torchaudio librosa
!pip install -q einops  # Necessário para operações de reshape eficientes

import os
import numpy as np
import pandas as pd
import torch
import torchaudio
import torch.nn as nn
import torch.nn.functional as F
import librosa
import librosa.display
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import KFold # Importando KFold para cross-validation
from tqdm import tqdm
from torchvision import transforms
from torchaudio.transforms import MelSpectrogram, AmplitudeToDB
from torch.optim.lr_scheduler import OneCycleLR
from google.colab import drive
import random
from einops import rearrange # Para operações de reshape complexas

# BLOCO 1.1: Semente para reprodutibilidade
def seed_everything(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False # Define como False para reprodutibilidade
seed_everything()

# Montar Google Drive
drive.mount('/content/drive')

# BLOCO 2: Definição de caminhos e hiperparâmetros
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"✅ Dispositivo selecionado: {device}")

base_path = "/content/drive/MyDrive/Mestrado/Curso/1o Semestre 2025/DeepLearning/ESC-50_dataset"
audio_path = os.path.join(base_path, "audio")
csv_path = os.path.join(base_path, "meta", "esc50.csv")
model_save_path = os.path.join(base_path, "modelos_ESC50")
os.makedirs(model_save_path, exist_ok=True)

# Hiperparâmetros
SAMPLE_RATE = 44100 # Conforme artigo
N_FFT = 1024        # Conforme artigo
HOP_LENGTH = 512    # Conforme artigo
N_MELS = 128        # Conforme artigo (mel-bands)
DURATION = 5        # segundos (5 segundos)
NUM_CLASSES = 50
BATCH_SIZE = 32     # Conforme artigo
NUM_EPOCHS = 100    # Podemos ajustar com EarlyStopping
LEARNING_RATE = 1e-3 # LR máximo para OneCycleLR

# K-Fold Cross-Validation
N_SPLITS = 5 # Conforme o dataset ESC-50 e o artigo

# BLOC0 3: Classe para Label Smoothing Cross Entropy
class LabelSmoothingCrossEntropy(nn.Module):
    def __init__(self, epsilon: float = 0.1, reduction='mean'):
        super().__init__()
        self.epsilon = epsilon
        self.reduction = reduction

    def forward(self, preds, target):
        n = preds.size()[-1]
        log_preds = F.log_softmax(preds, dim=-1)
        # Atenção: Esta parte da loss (soma sobre todas as classes) pode gerar valores altos,
        # mas é a formulação para label smoothing (sum over all true and smoothed labels).
        # A acurácia é o que realmente importa para a performance.
        loss = (-log_preds).sum(dim=-1)
        nll_loss = F.nll_loss(log_preds, target, reduction=self.reduction)
        return (1 - self.epsilon) * nll_loss + self.epsilon * loss.mean()

# BLOCO 4: Data Augmentation (SpecAugment)
class SpecAugment(nn.Module):
    def __init__(self, freq_mask_param=24, time_mask_param=48, p_time_warp=0.5):
        super().__init__()
        self.spec_augment = nn.Sequential(
            torchaudio.transforms.FrequencyMasking(freq_mask_param),
            torchaudio.transforms.TimeMasking(time_mask_param),
        )
        self.p_time_warp = p_time_warp # Probabilidade de aplicar time warping

    def forward(self, x):
        return self.spec_augment(x)

# BLOCO 5: Definição do Dataset e Dataloader
class AudioDataset(Dataset):
    def __init__(self, df, audio_dir, sample_rate, n_fft, hop_length, n_mels, duration,
                 transform=None, train=True, mean=None, std=None, target_mel_frames=216):
        self.annotations = df # DataFrame já filtrado para treino/validação
        self.audio_dir = audio_dir
        self.sample_rate = sample_rate
        self.n_fft = n_fft
        self.hop_length = hop_length
        self.n_mels = n_mels
        self.duration = duration
        self.transform = transform
        self.train = train
        self.target_mel_frames = target_mel_frames

        self.mel_spectrogram_transform = MelSpectrogram(
            sample_rate=self.sample_rate,
            n_fft=self.n_fft,
            hop_length=self.hop_length,
            n_mels=self.n_mels,
            normalized=False
        )
        self.amplitude_to_db_transform = AmplitudeToDB()

        # Valores MEAN/STD pré-calculados para ESC-50
        self.mean = mean if mean is not None else -46.68
        self.std = std if std is not None else 23.35

    def __len__(self):
        return len(self.annotations)

    def __getitem__(self, index):
        audio_path = os.path.join(self.audio_dir, self.annotations.iloc[index, 0])
        label = self.annotations.iloc[index, 4]

        waveform, sr = torchaudio.load(audio_path)

        if sr != self.sample_rate:
            waveform = torchaudio.transforms.Resample(orig_freq=sr, new_freq=self.sample_rate)(waveform)
            sr = self.sample_rate

        target_waveform_length = self.sample_rate * self.duration
        if waveform.shape[1] < target_waveform_length:
            padding = target_waveform_length - waveform.shape[1]
            waveform = F.pad(waveform, (0, padding))
        elif waveform.shape[1] > target_waveform_length:
            waveform = waveform[:, :target_waveform_length]

        mel_spec = self.mel_spectrogram_transform(waveform)
        mel_spec = self.amplitude_to_db_transform(mel_spec)

        if mel_spec.shape[0] > 1:
            mel_spec = torch.mean(mel_spec, dim=0, keepdim=True)

        # Ajuste para o número de frames do artigo (216)
        if mel_spec.shape[2] < self.target_mel_frames:
            padding = self.target_mel_frames - mel_spec.shape[2]
            mel_spec = F.pad(mel_spec, (0, padding))
        elif mel_spec.shape[2] > self.target_mel_frames:
            mel_spec = mel_spec[:, :, :self.target_mel_frames]

        mel_spec = (mel_spec - self.mean) / (self.std + 1e-6)

        if self.transform and self.train:
            mel_spec = self.transform(mel_spec)

        # Garante que o label seja um inteiro e do tipo long
        return mel_spec, torch.tensor(int(label), dtype=torch.long) # Correção aplicada aqui

# BLOCO 6: Implementação da Arquitetura RACNN (Conforme Artigo)
class RACModule(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(RACModule, self).__init__()
        self.dwconv = nn.Conv2d(in_channels, in_channels, kernel_size=3, padding=1, groups=in_channels, bias=False)
        self.bn_dw = nn.BatchNorm2d(in_channels)
        self.pwconv = nn.Conv2d(in_channels, out_channels, kernel_size=1, bias=False)
        self.bn_pw = nn.BatchNorm2d(out_channels)
        self.global_avg_pool = nn.AdaptiveAvgPool2d(1)
        self.fc1 = nn.Conv2d(out_channels, out_channels // 16, kernel_size=1, bias=False)
        self.relu = nn.ReLU(inplace=True)
        self.fc2 = nn.Conv2d(out_channels // 16, out_channels, kernel_size=1, bias=False)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.dwconv(x)
        x = self.bn_dw(x)
        x = F.relu(x, inplace=True)

        x = self.pwconv(x)
        x = self.bn_pw(x)
        out = F.relu(x, inplace=True)

        scale = self.global_avg_pool(out)
        scale = self.fc1(scale)
        scale = self.relu(scale)
        scale = self.fc2(scale)
        scale = self.sigmoid(scale)

        out = out * scale.expand_as(out)
        return out

class RACBlock(nn.Module):
    def __init__(self, in_channels, out_channels, stride=1, num_rac_modules=1):
        super(RACBlock, self).__init__()
        layers = []
        layers.append(RACModule(in_channels, out_channels))
        for _ in range(1, num_rac_modules):
             layers.append(RACModule(out_channels, out_channels))
        self.rac_modules = nn.Sequential(*layers)

        if stride == 2:
            self.pool = nn.AvgPool2d(kernel_size=2, stride=2)
        else:
            self.pool = nn.Identity()

    def forward(self, x):
        x = self.rac_modules(x)
        x = self.pool(x)
        return x

class RACNN(nn.Module):
    def __init__(self, num_classes=50):
        super(RACNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(32)
        self.relu1 = nn.ReLU(inplace=True)
        self.pool1 = nn.AvgPool2d(kernel_size=2, stride=2)

        self.rac_block2 = nn.Sequential(
            RACModule(32, 64),
            RACModule(64, 64)
        )
        self.pool2 = nn.AvgPool2d(kernel_size=2, stride=2)

        self.rac_block3 = nn.Sequential(
            RACModule(64, 128),
            RACModule(128, 128)
        )
        self.pool3 = nn.AvgPool2d(kernel_size=2, stride=2)

        self.rac_block4 = nn.Sequential(
            RACModule(128, 256),
            RACModule(256, 256)
        )
        self.pool4 = nn.AvgPool2d(kernel_size=2, stride=2)

        self.global_avg_pool = nn.AdaptiveAvgPool2d(1)

        self.fc1 = nn.Linear(256, 512)
        self.bn_fc1 = nn.BatchNorm1d(512)
        self.dropout1 = nn.Dropout(0.5)

        self.fc2 = nn.Linear(512, num_classes)

    def forward(self, x):
        x = self.conv1(x)
        x = self.bn1(x)
        x = self.relu1(x)
        x = self.pool1(x)

        x = self.rac_block2(x)
        x = self.pool2(x)

        x = self.rac_block3(x)
        x = self.pool3(x)

        x = self.rac_block4(x)
        x = self.pool4(x)

        x = self.global_avg_pool(x)
        x = torch.flatten(x, 1)

        x = self.fc1(x)
        x = self.bn_fc1(x)
        x = F.relu(x, inplace=True)
        x = self.dropout1(x)

        x = self.fc2(x)

        return x

# BLOCO 7: Treinamento e Avaliação com K-Fold Cross-Validation
class EarlyStopper:
    def __init__(self, patience=5, min_delta=0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.min_validation_acc = 0.0
        self.early_stop = False

    def step(self, validation_acc):
        if validation_acc > self.min_validation_acc + self.min_delta:
            self.min_validation_acc = validation_acc
            self.counter = 0
        else:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
                print(f"Paciência esgotada ({self.patience} épocas sem melhoria).")

# Carregar o dataset completo
esc50_df = pd.read_csv(csv_path)

# Inicializar K-Fold Cross-Validation
kf = KFold(n_splits=N_SPLITS, shuffle=True, random_state=42)

all_fold_accuracies = []
all_fold_train_losses = []
all_fold_val_losses = []
all_fold_train_accuracies = []
all_fold_val_accuracies = []

print(f"\n🚀 Iniciando o Treinamento da RACNN com {N_SPLITS}-Fold Cross-Validation...")

for fold, (train_index, val_index) in enumerate(kf.split(esc50_df), 1):
    print(f"\n==================== Iniciando Fold {fold}/{N_SPLITS} ====================")
    best_model_fold_path = os.path.join(model_save_path, f"best_racnn_article_ESC50_fold{fold}.pt")

    train_df = esc50_df.iloc[train_index]
    val_df = esc50_df.iloc[val_index]

    # Instanciar o SpecAugment para o treino
    train_transform = SpecAugment(freq_mask_param=24, time_mask_param=48)

    train_dataset = AudioDataset(
        df=train_df,
        audio_dir=audio_path,
        sample_rate=SAMPLE_RATE,
        n_fft=N_FFT,
        hop_length=HOP_LENGTH,
        n_mels=N_MELS,
        duration=DURATION,
        transform=train_transform,
        train=True
    )

    val_dataset = AudioDataset(
        df=val_df,
        audio_dir=audio_path,
        sample_rate=SAMPLE_RATE,
        n_fft=N_FFT,
        hop_length=HOP_LENGTH,
        n_mels=N_MELS,
        duration=DURATION,
        transform=None,
        train=False
    )

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

    # Reiniciar o modelo e otimizadores para cada fold
    model = RACNN(num_classes=NUM_CLASSES).to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)
    criterion = LabelSmoothingCrossEntropy(epsilon=0.1)

    scheduler = OneCycleLR(optimizer,
                           max_lr=LEARNING_RATE * 10,
                           steps_per_epoch=len(train_loader),
                           epochs=NUM_EPOCHS,
                           anneal_strategy='cos',
                           pct_start=0.3,
                           div_factor=25.0,
                           final_div_factor=10000.0)

    early_stopper = EarlyStopper(patience=10, min_delta=0.001)

    fold_train_loss_history = []
    fold_train_acc_history = []
    fold_val_loss_history = []
    fold_val_acc_history = []
    best_val_acc_fold = 0.0

    for epoch in range(1, NUM_EPOCHS + 1):
        model.train()
        total_loss = 0
        correct = 0
        total = 0
        progress_bar = tqdm(train_loader, desc=f"📘 Fold {fold} - Época {epoch}/{NUM_EPOCHS}")

        for batch_idx, (X, y) in enumerate(progress_bar):
            X, y = X.to(device), y.to(device).long()

            optimizer.zero_grad()
            outputs = model(X)
            loss = criterion(outputs, y)
            loss.backward()
            optimizer.step()
            scheduler.step()

            total_loss += loss.item() * X.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += y.size(0)
            correct += (predicted == y).sum().item()

            current_lr = optimizer.param_groups[0]['lr']
            progress_bar.set_postfix(loss=loss.item(), acc=correct/total, lr=f"{current_lr:.6f}")

        train_epoch_loss = total_loss / total
        train_epoch_acc = correct / total
        fold_train_loss_history.append(train_epoch_loss)
        fold_train_acc_history.append(train_epoch_acc)

        # Validação
        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for X, y in val_loader:
                X, y = X.to(device), y.to(device).long()
                outputs = model(X)
                loss = criterion(outputs, y)

                val_loss += loss.item() * X.size(0)
                _, preds = torch.max(outputs, 1)
                val_correct += (preds == y).sum().item()
                val_total += y.size(0)

        val_epoch_loss = val_loss / val_total
        val_epoch_acc = val_correct / val_total
        fold_val_loss_history.append(val_epoch_loss)
        fold_val_acc_history.append(val_epoch_acc)

        # Novo log aprimorado
        log_message = (
            f"Validação - Loss: {val_epoch_loss:.4f} | Acurácia: {val_epoch_acc:.4f} "
            f"| LR: {current_lr:.6f}"
        )

        if val_epoch_acc > best_val_acc_fold:
            best_val_acc_fold = val_epoch_acc
            torch.save(model.state_dict(), best_model_fold_path)
            log_message += f" | ✅ Novo Melhor da Fold {fold} ({best_val_acc_fold:.4f}) salvo!"
        print(log_message)

        # Early Stopping
        early_stopper.step(val_epoch_acc)
        if early_stopper.early_stop:
            print(f"\n⏹️ Early stopping ativado para Fold {fold}. Melhor Acurácia da Fold: {best_val_acc_fold:.4f}")
            break

    all_fold_accuracies.append(best_val_acc_fold)
    all_fold_train_losses.append(fold_train_loss_history)
    all_fold_val_losses.append(fold_val_loss_history)
    all_fold_train_accuracies.append(fold_train_acc_history)
    all_fold_val_accuracies.append(fold_val_acc_history)

# BLOCO 8: Resultados Finais do K-Fold e Gráficos Agregados
print("\n==================== Treinamento K-Fold Concluído ====================")
print(f"📊 Acurácias de Validação por Fold: {all_fold_accuracies}")
print(f"Média das Acurácias de Validação (K-Fold): {np.mean(all_fold_accuracies):.4f}")
print(f"Desvio Padrão das Acurácias de Validação (K-Fold): {np.std(all_fold_accuracies):.4f}")

# Plotar gráficos agregados (opcional: média de todas as folds ou primeira fold)
# Para simplificar a visualização, podemos plotar os gráficos da primeira fold
# ou calcular a média das curvas se todas as folds rodarem pelo mesmo número de épocas.
# Para este exemplo, vamos plotar a primeira fold. Para uma visualização mais complexa,
# você pode calcular a média das curvas de loss/acc ao longo das épocas.

if len(all_fold_train_losses) > 0:
    print("\n📈 Gerando gráficos de desempenho da primeira Fold (exemplo visual)...")
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(all_fold_train_losses[0], label='Treino (Fold 1)')
    plt.plot(all_fold_val_losses[0], label='Validação (Fold 1)')
    plt.title('Loss por Época (Primeira Fold)')
    plt.xlabel('Épocas')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(all_fold_train_accuracies[0], label='Treino (Fold 1)')
    plt.plot(all_fold_val_accuracies[0], label='Validação (Fold 1)')
    plt.title('Acurácia por Época (Primeira Fold)')
    plt.xlabel('Épocas')
    plt.ylabel('Acurácia')
    plt.legend()

    plt.tight_layout()
    plt.show()

print(f"\n✨ Treinamento K-Fold concluído! Modelos salvos na pasta: {model_save_path}")
```
