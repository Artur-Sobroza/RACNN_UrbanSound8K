```python
"""
Script para uma an√°lise detalhada e multifacetada do dataset UrbanSound8K.
Objetivos:
1.  Verificar a integridade e qualidade (sil√™ncio/dura√ß√£o) de *TODOS* os arquivos de √°udio
    em *TODAS* as classes.
2.  Gerar 10 Mel-Espectrogramas para a classe 'street_music' para an√°lise visual aprofundada
    da sua heterogeneidade.
3.  Fornecer dados e visualiza√ß√µes para uma nova se√ß√£o do artigo focada nas caracter√≠sticas
    intr√≠nsecas do UrbanSound8K.
"""

# BLOCO 1: Instala√ß√£o de bibliotecas e importa√ß√µes
# --------------------------------------------------------------------------------
!pip install -q torchaudio librosa matplotlib pandas seaborn
!pip install -q soundfile # Para verificar arquivos WAV

import os
import numpy as np
import pandas as pd
import torch
import torchaudio
import torch.nn.functional as F # Importar F para F.pad
import librosa
import librosa.display
import matplotlib.pyplot as plt
import soundfile as sf # Para carregar WAV e verificar integridade
from google.colab import drive
import random
from tqdm import tqdm

# BLOCO 2: Montar Google Drive e Definir Caminhos
# --------------------------------------------------------------------------------
drive.mount('/content/drive')

# Caminhos para o Dataset UrbanSound8K
base_path_us8k = "/content/drive/MyDrive/Mestrado/Curso/1o Semestre 2025/DeepLearning/UrbanSound8K_dataset"
csv_path_us8k = os.path.join(base_path_us8k, "metadata", "UrbanSound8K.csv")
audio_base_path_us8k = os.path.join(base_path_us8k, "audio")

# Semente para reprodutibilidade
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)

print("‚úÖ Setup inicial conclu√≠do. Caminhos definidos.")

# BLOCO 3: Carregar Metadados e Configura√ß√µes Globais
# --------------------------------------------------------------------------------
print(f"\n‚è≥ Carregando metadados do UrbanSound8K de: {csv_path_us8k}")
try:
    us8k_df = pd.read_csv(csv_path_us8k)
    print(f"‚úÖ Metadados carregados. Total de amostras: {len(us8k_df)}")
except FileNotFoundError:
    print(f"‚ùå ERRO: Arquivo CSV do UrbanSound8K n√£o encontrado em {csv_path_us8k}")
    print("Por favor, verifique o caminho e a montagem do Google Drive.")
    exit()

unique_classes = us8k_df['class'].unique()
print(f"N√∫mero de classes √∫nicas no UrbanSound8K: {len(unique_classes)}")
print(f"Nomes das classes UrbanSound8K: {list(unique_classes)}")

# Par√¢metros para verifica√ß√£o de √°udio e Mel-espectrograma, consistentes com o treinamento
SAMPLE_RATE = 44100
MIN_DURATION_SEC = 0.5 # Dura√ß√£o m√≠nima para √°udio v√°lido
AMPLITUDE_THRESHOLD = 1e-4 # Limiar de amplitude para considerar √°udio como n√£o-silencioso
N_FFT = 1024
HOP_LENGTH = 512
N_MELS = 128
DURATION_US8K = 4 # segundos (dura√ß√£o padr√£o dos √°udios processados pelo modelo)
TARGET_MEL_FRAMES = 216 # Conforme usado no modelo RACNN

print("\n‚öôÔ∏è Configura√ß√µes para an√°lise: ")
print(f"  - Sample Rate: {SAMPLE_RATE}")
print(f"  - Min. Dura√ß√£o p/ √Åudio V√°lido: {MIN_DURATION_SEC}s")
print(f"  - Limiar de Amplitude p/ N√£o-Silencioso: {AMPLITUDE_THRESHOLD}")
print(f"  - N_MELS (Mel-Espectrogramas): {N_MELS}")

# BLOCO 4: Verifica√ß√£o de Integridade e Conte√∫do para *TODAS* as Classes
# --------------------------------------------------------------------------------
print("\nüîç Iniciando verifica√ß√£o de integridade e conte√∫do para *TODAS* as classes...")

# Dicion√°rio para armazenar os resultados da verifica√ß√£o por classe
class_integrity_results = {}
all_problematic_files_detailed = [] # Para armazenar todos os arquivos problem√°ticos, com classe e raz√£o

# Fun√ß√£o auxiliar para verifica√ß√£o (repetida para clareza e independ√™ncia)
def is_silent_or_too_short(waveform, sr, min_duration_sec, amplitude_threshold):
    """
    Verifica se um waveform √© considerado 'silencioso' ou muito curto.
    waveform: tensor do √°udio (espera-se que j√° esteja na CPU)
    sr: sample rate
    min_duration_sec: dura√ß√£o m√≠nima em segundos para considerar um √°udio v√°lido
    amplitude_threshold: Limiar de amplitude para considerar um √°udio como n√£o-silencioso
    """
    if waveform.numel() == 0: # Checa se o tensor est√° vazio
        return True, "vazio"

    duration = waveform.shape[-1] / sr # Dura√ß√£o em segundos (considera mono ou est√©reo, pega o √∫ltimo dim)

    if duration < min_duration_sec:
        return True, f"muito curto ({duration:.2f}s)"

    waveform_abs = waveform.abs().cpu()
    if waveform_abs.ndim > 1: # Se multi-canal, achata para uma dimens√£o para o max
        waveform_abs = waveform_abs.flatten()
    max_amplitude = torch.max(waveform_abs).item()

    if max_amplitude < amplitude_threshold:
        return True, "silencioso (abaixo do limiar de amplitude)"

    return False, "ok"

# Iterar sobre cada classe √∫nica
for class_name in unique_classes:
    print(f"\n--- Verificando classe: '{class_name}' ---")
    class_df = us8k_df[us8k_df['class'] == class_name]
    total_samples_in_class = len(class_df)
    problematic_count = 0
    problem_reasons_counts = {} # Para contar os tipos de problemas (silencioso, muito curto, erro)

    for index, row in tqdm(class_df.iterrows(), total=total_samples_in_class, desc=f"Processando {class_name}"):
        file_name = row['slice_file_name']
        fold_num = row['fold']
        audio_path = os.path.join(audio_base_path_us8k, f"fold{fold_num}", file_name)

        is_prob = False
        reason = "ok"

        try:
            waveform, sr = torchaudio.load(audio_path)
            if waveform.ndim > 1 and waveform.shape[0] > 1:
                waveform = torch.mean(waveform, dim=0, keepdim=True) # Converte para mono

            is_prob, reason = is_silent_or_too_short(waveform, sr, MIN_DURATION_SEC, AMPLITUDE_THRESHOLD)

            if is_prob:
                problematic_count += 1
                problem_reasons_counts[reason] = problem_reasons_counts.get(reason, 0) + 1
                all_problematic_files_detailed.append({'class': class_name, 'file_path': audio_path, 'reason': reason})

        except Exception as e:
            is_prob = True
            reason = f"erro ao carregar: {e}"
            problematic_count += 1
            problem_reasons_counts[reason] = problem_reasons_counts.get(reason, 0) + 1
            all_problematic_files_detailed.append({'class': class_name, 'file_path': audio_path, 'reason': reason})

    class_integrity_results[class_name] = {
        'total_samples': total_samples_in_class,
        'problematic_count': problematic_count,
        'problem_reasons': problem_reasons_counts
    }
    print(f"  Total de amostras: {total_samples_in_class}")
    print(f"  Arquivos problem√°ticos/silenciosos: {problematic_count}")
    if problematic_count > 0:
        print(f"  Detalhes dos problemas: {problem_reasons_counts}")

print("\n--- RESUMO GERAL DA VERIFICA√á√ÉO DE INTEGRIDADE DOS DADOS ---")
integrity_summary_df = pd.DataFrame.from_dict(class_integrity_results, orient='index')
integrity_summary_df['problematic_percentage'] = (integrity_summary_df['problematic_count'] / integrity_summary_df['total_samples']) * 100
print(integrity_summary_df.sort_values(by='problematic_percentage', ascending=False))

if all_problematic_files_detailed:
    print("\nDetalhes completos de TODOS os arquivos problem√°ticos encontrados:")
    for item in all_problematic_files_detailed:
        print(f"  - Classe: {item['class']}, Arquivo: {item['file_path']}, Raz√£o: {item['reason']}")
else:
    print("\nNenhum arquivo problem√°tico/silencioso encontrado em nenhuma classe do dataset completo (com base nos limiares definidos).")


# BLOCO 5: Visualiza√ß√£o de 10 Mel-Espectrogramas da Classe `street_music`
# --------------------------------------------------------------------------------
print("\nüìä Gerando 10 Mel-Espectrogramas para amostras de 'street_music'...")

# Fun√ß√£o para gerar e plotar Mel-espectrogramas (reusada, com pequena adapta√ß√£o)
def plot_mel_spectrogram(audio_path, title="Mel-Espectrograma", ax=None):
    try:
        waveform, sr = torchaudio.load(audio_path)
        if sr != SAMPLE_RATE:
            waveform = torchaudio.transforms.Resample(orig_freq=sr, new_freq=SAMPLE_RATE)(waveform)
            sr = SAMPLE_RATE

        # Padroniza√ß√£o da Dura√ß√£o do √Åudio (para 4 segundos)
        target_waveform_length = SAMPLE_RATE * DURATION_US8K
        if waveform.shape[1] < target_waveform_length:
            padding = target_waveform_length - waveform.shape[1]
            waveform = F.pad(waveform, (0, padding))
        elif waveform.shape[1] > target_waveform_length:
            waveform = waveform[:, :target_waveform_length]

        if waveform.shape[0] > 1: # Converte para mono
            waveform = torch.mean(waveform, dim=0, keepdim=True)

        mel_spectrogram_transform = torchaudio.transforms.MelSpectrogram(
            sample_rate=SAMPLE_RATE, n_fft=N_FFT, hop_length=HOP_LENGTH, n_mels=N_MELS
        )
        amplitude_to_db_transform = torchaudio.transforms.AmplitudeToDB()

        mel_spec = mel_spectrogram_transform(waveform)
        mel_spec_db = amplitude_to_db_transform(mel_spec)

        # Padroniza√ß√£o dos frames (216)
        if mel_spec_db.shape[2] < TARGET_MEL_FRAMES:
            padding_frames = TARGET_MEL_FRAMES - mel_spec_db.shape[2]
            mel_spec_db = F.pad(mel_spec_db, (0, padding_frames))
        elif mel_spec_db.shape[2] > TARGET_MEL_FRAMES:
            mel_spec_db = mel_spec_db[:, :, :TARGET_MEL_FRAMES]

        if ax is None:
            fig, ax = plt.subplots(figsize=(10, 4))
        else:
            fig = ax.figure

        im = librosa.display.specshow(mel_spec_db.squeeze().cpu().numpy(),
                                 sr=SAMPLE_RATE, hop_length=HOP_LENGTH, x_axis='time', y_axis='mel', ax=ax)
        ax.set_title(title)
        fig.colorbar(im, ax=ax, format='%+2.0f dB')
        ax.set_xlabel("Tempo (s)")
        ax.set_ylabel("Frequ√™ncia Mel")
        return True
    except Exception as e:
        if ax is not None:
            ax.text(0.5, 0.5, f"Erro ao plotar: {e}\n{os.path.basename(audio_path)}",
                    horizontalalignment='center', verticalalignment='center',
                    transform=ax.transAxes, color='red', fontsize=12)
            ax.set_title(f"Erro para {os.path.basename(audio_path)}")
        else:
            print(f"‚ùå Erro ao processar/plotar {audio_path}: {e}")
        return False

# Filtrar apenas as amostras da classe 'street_music'
street_music_df = us8k_df[us8k_df['class'] == 'street_music']

# Excluir arquivos j√° identificados como problem√°ticos na verifica√ß√£o completa
# para ter uma vis√£o mais clara dos √°udios "v√°lidos" mas heterog√™neos.
problematic_street_music_paths = [item['file_path'] for item in all_problematic_files_detailed if item['class'] == 'street_music']
valid_street_music_df = street_music_df[~street_music_df.apply(lambda row: os.path.join(audio_base_path_us8k, f"fold{row['fold']}", row['slice_file_name']) in problematic_street_music_paths, axis=1)]


if len(valid_street_music_df) > 0:
    num_plots = min(10, len(valid_street_music_df)) # Agora 10 plots
    sample_to_plot_df = valid_street_music_df.sample(n=num_plots, random_state=42)

    fig, axes = plt.subplots(num_plots, 1, figsize=(12, 4 * num_plots))
    if num_plots == 1:
        axes = [axes]

    for i, (index, row) in enumerate(sample_to_plot_df.iterrows()):
        file_name = row['slice_file_name']
        fold_num = row['fold']
        audio_path = os.path.join(audio_base_path_us8k, f"fold{fold_num}", file_name)
        plot_mel_spectrogram(audio_path, title=f"Mel-Espectrograma de street_music - {file_name}", ax=axes[i])
    plt.tight_layout()
    plt.show()
else:
    print("N√£o h√° amostras 'street_music' v√°lidas suficientes para plotar espectrogramas ap√≥s filtragem.")

# BLOCO 6: Sum√°rio Final para o Artigo
# --------------------------------------------------------------------------------
print("\n--- Sum√°rio para a Se√ß√£o 'An√°lise Detalhada das Caracter√≠sticas do Dataset UrbanSound8K' ---")
print("\n**1. Balan√ßo de Classes:**")
print("Apresentar a contagem de amostras por classe (total do dataset) e os gr√°ficos de barra.")
print(us8k_df['class'].value_counts())

print("\n**2. Integridade e Qualidade dos Arquivos de √Åudio:**")
print("Descrever a metodologia de verifica√ß√£o de arquivos (corrompidos, silenciosos, curtos).")
print("Incluir a tabela de resumo de integridade por classe:")
print(integrity_summary_df.sort_values(by='problematic_percentage', ascending=False))
print("Se houver arquivos problem√°ticos, detalhar alguns exemplos.")

print("\n**3. Estudo de Caso: Heterogeneidade Ac√∫stica da Classe 'street_music':**")
print("Mencionar o resultado de 0 arquivos problem√°ticos para 'street_music' (se for o caso).")
print("Discutir a extrema heterogeneidade visual observada nos 10 Mel-espectrogramas gerados (incluir as imagens no artigo).")
print("Enfatizar a aus√™ncia de um padr√£o ac√∫stico consistente para esta classe.")
print("Concluir que o problema n√£o √© a integridade dos dados, mas a falta de coer√™ncia do conte√∫do rotulado.")

print("\nEstas an√°lises fundamentam a discuss√£o sobre o desempenho do modelo e os desafios da generaliza√ß√£o no UrbanSound8K.")
```
